```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
matplotlib.rcParams['font.family'] = 'appleGothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('./data/중간시험/서울시 기간별 일평균 대기환경 정보.csv', encoding='cp949')
df['측정일자'] = pd.to_datetime(df['측정일자'], format='%Y%m%d')
df.rename(columns={'미세먼지(㎍/㎥)': '미세먼지', '초미세먼지(㎍/㎥)': '초미세먼지', '오존(ppm)': '오존'}, inplace=True)
```

```python
ddg = df[df['측정소명'] == '동대문구']
ddg1 = ddg.query("(측정일자 >= '2023-01-01') & (측정일자 <= '2023-03-31')")
ddg1
```

```python
plt.figure(figsize=(12, 6))
plt.title('김태훈 1번 미세먼지 그래프')
plt.plot(ddg1['측정일자'], ddg1['미세먼지'], label='미세먼지 농도', color='red', marker='o', linestyle=':')
plt.legend()
```

```python
plt.show()
plt.savefig('./data/중간시험/김태훈_1번_미세먼지_그래프.png', dpi=150)
```

```python
ddg_current_bad = ddg1.query("미세먼지 >= 51")
ddg_revised_bad = ddg1.query("미세먼지 >= 36")
ddg_day_count = {'현행 미세먼지 예보기준 나쁨': len(ddg_current_bad), '개정 미세먼지 예보기준 나쁨': len(ddg_revised_bad)}
ddg_day_count
```

```python
names = list(ddg_day_count.keys())
values = list(ddg_day_count.values())
plt.bar(names, values)
plt.ylabel('미세먼지 예보 기준 나쁨 이상의 날')
plt.title('김태훈 2번 미세먼지 그래프')
```

```python
ddg_2023 = ddg.query("(측정일자 >= '2023-01-01') & (측정일자 <= '2023-12-31')")
ddg_2023
```

```python
jrg = df[df['측정소명'] == '중랑구']
jrg
jrg_2023 = jrg.query("(측정일자 >= '2023-01-01') & (측정일자 <= '2023-12-31')")
jrg_2023
```

```python
plt.figure(figsize=(12, 6))
plt.plot(ddg_2023['측정일자'], ddg_2023['미세먼지'], label='동대문구 미세먼지 농도', color='red')
plt.plot(jrg_2023['측정일자'], jrg_2023['미세먼지'], label='중랑구 미세먼지 농도', color='blue')
plt.xlabel('측정일자')
plt.ylabel('미세먼지')
plt.xticks(rotation=45)
plt.legend(ncol=2)
plt.title('김태훈 3번 미세먼지 그래프')
```

```python
ddg_mean = ddg_2023['미세먼지'].mean()
jrg_mean = jrg_2023['미세먼지'].mean()
means = {'동대문구 미세먼지 평균': ddg_mean, '중랑구 미세먼지 평균': jrg_mean}
means
```

```python
mean_names = list(means.keys())
mean_values = list(means.values())
plt.bar(mean_names, mean_values)
plt.ylabel('미세먼지 평균')
plt.title('김태훈 4번 미세먼지 그래프')
```

```python
from scipy.stats import pearsonr
corr1, p_value1 = pearsonr(df['미세먼지'], df['초미세먼지'])
```

```python
print(f"피어슨 상관계수: {corr1}")
print(f"p-value: {p_value1}")

if p_value1 < 0.05:
    print("상관관계가 통계적으로 유의미합니다.")
else:
    print("상관관계가 통계적으로 유의미하지 않습니다.")
```

```python
corr2, p_value2 = pearsonr(df['미세먼지'], df['오존'])
```

```python
print(f"피어슨 상관계수: {corr2}")
print(f"p-value: {p_value2}")

if p_value2 < 0.05:
    print("상관관계가 통계적으로 유의미합니다.")
else:
    print("상관관계가 통계적으로 유의미하지 않습니다.")
```
