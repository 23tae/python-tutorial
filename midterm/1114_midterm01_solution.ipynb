```python
import pandas as pd
from scipy.stats import pearsonr

corr, p_value = pearsonr(data['미세먼지(㎍/㎥)'], data['초미세먼지(㎍/㎥)'])

print(f"피어슨 상관계수: {corr}")
print(f"p-value: {p_value}")

# 만약 p-value가 0.05보다 작으면 상관관계가 통계적으로 유의미하다고 할 수 있습니다.
if p_value < 0.05:
    print("상관관계가 통계적으로 유의미합니다.")
else:
    print("상관관계가 통계적으로 유의미하지 않습니다.")
```

```python
import pandas as pd
from scipy.stats import pearsonr

corr, p_value = pearsonr(data['미세먼지(㎍/㎥)'], data['오존(ppm)'])

print(f"피어슨 상관계수: {corr}")
print(f"p-value: {p_value}")

# 만약 p-value가 0.05보다 작으면 상관관계가 통계적으로 유의미하다고 할 수 있습니다.
if p_value < 0.05:
    print("상관관계가 통계적으로 유의미합니다.")
else:
    print("상관관계가 통계적으로 유의미하지 않습니다.")
```
