# 텍스트 삽입

```python
plt.plot(x, y, marker='o')

for idx, txt in enumerate(y):
    plt.text(x[idx], y[idx], txt)
```

```python
plt.plot(x, y, marker='o')

for idx, txt in enumerate(y):
    plt.text(x[idx], y[idx] + 0.3, txt, ha='center', color ='blue')
```

```python
days = [1, 2, 3]
tmp = [28 , 29, 30]
rh = [34, 30, 36]
HI = [31, 29.5, 33]

plt.plot(days, tmp)
plt.plot(days, rh)
plt.plot(days, HI)
```

```python
plt.plot(days, tmp, label ='tmp')
plt.plot(days, rh, label = 'rh', marker = 'o', ls ='--')
plt.plot(days, HI, label = 'HI', marker = 's', ls ='-.')
plt.legend()
```
