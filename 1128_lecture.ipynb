#### 2. shp파일 불러와서 데이터 프레임으로 다루기

```python
import geopandas as gpd
import pandas as pd
import numpy as np
```

```python
#shp 파일을 pandas의 dataframe 처럼 불러오는 방법
sig = gpd.read_file('./data/sig_20230729/sig.shp', encoding = 'utf-8')
sig.head()
```

```python
#변수 sig의 타입을 살펴보자
type(sig)
```

GeoDataFrame은 "Geometry" 컬럼을 가지고 있으며, 이는 일반적으로 공간을 다루고 있는 정보의 데이터 타입을 저장하고 있음   



```python
#데이터를 시각화하고 싶을땐, plot() 함수를 사용하여 하자
sig.plot()
```

```python
#plot 함수는 color나 edgecolor, figsize등의 파라미터 설정이 가능하다
sig.plot(color='white', edgecolor = 'black')
```

```python
#GeoDataFrame은 Pandas의 데이터프레임처럼 다룰 수 있음.
#서울시 시군구 데이터만 추출해보자
seoul = sig[sig['SIG_CD'].str.startswith('11')]
seoul
```

```python
seoul.plot()
```

```python
#이번에는 point 자료를 불러와보자
ecosys_dot = gpd.read_file('./data/202211_생태계정밀조사_식물상_점/tbl_big_ecpe_flr_point.shp', encoding = 'utf-8')
ecosys_dot
```

```python
ecosys_dot.plot()
```

하지만, 이러한 point 데이터의 단순 시각화로는 어디에 데이터가 위치해있는지 알기 어렵기 때문에 위의 시군구 폴리곤 데이터등을 결합하여 이용해야한다.

```python
ax = sig.plot(color='white', edgecolor = 'black')
ecosys_dot.plot(ax=ax, color = 'navy')
```

지도가 이상하게 나오는 까닭은, 좌표계 설정이 제대로 되어있지 않아서 그렇다. 두 shp 파일의 좌표계를 확인해보자

```python
sig.crs
```

```python
ecosys_dot.crs
```

```python
sigCRS = sig.to_crs(epsg=4737)
ecosys_dotCRS = ecosys_dot.to_crs(epsg=4737)
```

```python
ax = sigCRS.plot(color='white',edgecolor='black', figsize=(7,7))
ecosys_dotCRS.plot (ax=ax, color='navy', figsize=(7,7))
```

```python
Chungbuk = sigCRS[sigCRS['SIG_CD'].str.startswith('43')]
Gyeongbuk = sigCRS[sigCRS['SIG_CD'].str.startswith('47')]
```

```python
CB_dot = lambda x:np.sum(ecosys_dotCRS.intersects(x))
```

```python
Chungbuk['count'] = Chungbuk['geometry'].apply(CB_dot)

Chungbuk
```
