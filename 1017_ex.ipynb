```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./data/1010_LST_실습데이터.csv', encoding='utf-8')
df = pd.DataFrame(data, columns=['날짜','평균기온','최고온도','최저기온'])
df
df['날짜'] = pd.to_datetime(df['날짜'], format='%Y%m%d')
```

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'appleGothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['날짜'], df['평균기온'], label='평균기온', color='orange', marker='o', linestyle='--')
plt.plot(df['날짜'], df['최고온도'], label='최고온도', color='red', marker='*', linestyle=':')
plt.plot(df['날짜'], df['최저기온'], label='최저기온', color='blue', marker='x', linestyle='-.')
```

```python
plt.title('김태훈 지표면 온도 데이터 시각화')
plt.xlabel('날짜')
plt.ylabel('기온 (°C)')
plt.legend(ncol=3, loc='lower left')
```

```python
plt.savefig('김태훈 지표면 온도 데이터 시각화.png')
plt.show()
```
