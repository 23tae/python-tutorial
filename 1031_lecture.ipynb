# 저번 시간 질문
1. 텍스트에서 x,y 둘다 표시하는법
2. 실습과제에서 x축을 시간 프레임으로 변경하는 법

```python
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [2, 4, 10]

plt.plot(x, y, marker='o')

for idx,(xi, yi) in enumerate(zip(x, y)):
    plt.text(xi, yi, f"({xi}, {yi})")  # x, y 좌표 둘 다 표시

# zip(x, y)는 x와 y 리스트의 각 원소를 짝지어서 튜플 형태로 반환
# f"({xi}, {yi})"는 f-string (formatted string literal)을 사용하여 xi와 yi 값을 문자열로 변환
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 한글폰트 추가하기 (맥북 버전)
import matplotlib.pyplot as plt
import pandas as pd
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #윈도우
csv_file = r'./data/1009_LST 구축 년월.csv'
df = pd.read_csv(csv_file)

# 날짜 데이터 type 변환
df['날짜'] = pd.to_datetime(df['날짜'], format='%Y%m%d')
df['날짜'].dtype

plt.figure(figsize=(18, 6))
plt.plot(df['날짜'], df['최고온도'], color='red', marker='o', markersize = 7, linestyle='-', label='최고온도')
plt.plot(df['날짜'], df['평균기온'], color='orange', marker='x', markersize = 7, linestyle='-', label='평균기온')
plt.plot(df['날짜'], df['최저기온'], color='blue', marker='^', markersize = 7, linestyle='-.', label='최저기온')

# 범례 표시
plt.legend(fontsize = 20, loc = "upper left")

# 그래프 제목
plt.title('LST', fontsize = 15)

# 그래프 x,y축 label 추가.
plt.xlabel('날짜', fontsize=17)
plt.ylabel('온도(°C)', fontsize=17)
#plt.grid(axis ='y', color='b', alpha = 0.5, linestyle = '--')
# 그래프 저장하기
#plt.savefig('지표면 온도 데이터 시각화.png', dpi = 200)
```

# 여러 그래프

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['font.family'] = 'Malgun Gothic' 
matplotlib.rcParams['axes.unicode_minus'] = False 
```

```python
df = pd.read_csv(r'./data/파이썬수업자료_1031.csv')
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15,10)) # 2 x 2 plot
fig.suptitle('여러 그래프')

axs[0,0].bar(df["수목명"],df['수고'], label='수고')
axs[0,0].set_title('첫번째그래프')
axs[0,0].legend(loc=(0.4,0.85))
axs[0,0].set(xlabel = '수목', ylabel='수고')
axs[0,0].set_xticklabels(df['수목명'], rotation = 90)
axs[0,0].set_facecolor('#2a9d8f')
axs[0,0].grid(linestyle='--', linewidth= 0.5)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15,10)) # 2 x 2 plot
fig.suptitle('여러 그래프')

axs[0,0].bar(df["수목명"],df['수고'], label='수고')
axs[0,0].set_title('첫번째그래프')
axs[0,0].legend(loc=(0.4,0.85))
axs[0,0].set(xlabel = '수목', ylabel='수고')
axs[0,0].set_facecolor('#2a9d8f')
axs[0,0].grid(linestyle='--', linewidth= 0.5)

#두번째 그래프
axs[0,1].plot(df['수목명'], df['수관폭'], label='수관폭')
axs[0,1].plot(df['수목명'], df['흉고직경'], label='흉고직경')
axs[0,1].legend()

#세번째 그래프
axs[1,0].barh(df["수목명"],df['수고'], label='수고')

#네번째 그래프
axs[1,1].scatter(df['수목명'], df['흉고직경'], label='흉고직경')
```
