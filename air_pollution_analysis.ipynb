```python
import pandas as pd
```

```python
bongdam_df = pd.read_csv('./data/bongdam_data.txt', encoding='utf-8', sep='\t')
```

```python
bongdam_df
```

```python
start_date = '2023-09-22'
end_date = '2023-11-21'
```

```python
def get_average_value(data):
  data['날짜(년-월-일)'] = pd.to_datetime(data['날짜(년-월-일)'])
  data.set_index('날짜(년-월-일)', inplace=True)
  data.sort_index(inplace=True)
  selected_rows = data.loc[start_date:end_date]
  average_values = selected_rows.mean()
  return average_values
```

```python
bongdam_average = get_average_value(bongdam_df)
```

```python
average_df = pd.DataFrame(bongdam_average).transpose()
average_df.reset_index(inplace=True)
average_df.rename(columns={'index': '지역'}, inplace=True)
average_df['지역'] = '봉담읍'
```

```python
average_df
```

```python
namyang_df = pd.read_csv('./data/namyang_data.txt', encoding='utf-8', sep='\t')
```

```python
namyang_df
```

```python
saesol_df = pd.read_csv('./data/saesol_data.txt', encoding='utf-8', sep='\t')
```

```python
saesol_df
```

```python
namyang_average = get_average_value(namyang_df)
saesol_average = get_average_value(saesol_df)
```

```python
row_namyang = pd.DataFrame(namyang_average).transpose()
row_namyang['지역'] = '남양읍'
row_saesol = pd.DataFrame(saesol_average).transpose()
row_saesol['지역'] = '새솔동'
```

```python
average_df = pd.concat([average_df, row_namyang])
average_df = pd.concat([average_df, row_saesol])
```

```python
average_df
```

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['font.family'] = 'AppleGothic' 
matplotlib.rcParams['axes.unicode_minus'] = False 
```

```python
n = average_df.shape[0]
index = np.arange(n)
```

```python
index
```

```python
w = 0.2
plt.figure(figsize=(16, 10))
plt.bar(index - w/2, average_df['PM-10(㎍/㎥)'], width=w, label='PM-10')
plt.bar(index + w/2, average_df['PM-2.5(㎍/㎥)'], width=w, label='PM-2.5')

for i, value in enumerate(average_df['PM-10(㎍/㎥)']):
    plt.text(i - w/2, value + 1, f'{value:.2f}', ha='center', va='bottom')

for i, value in enumerate(average_df['PM-2.5(㎍/㎥)']):
    plt.text(i + w/2, value + 1, f'{value:.2f}', ha='center', va='bottom')

plt.legend(ncol=2)
plt.xticks(index, average_df['지역'])
plt.title('화성시 미세먼지 평균')

plt.ylim(0, max(average_df[['PM-10(㎍/㎥)', 'PM-2.5(㎍/㎥)']].values.max() + 5, 20))

plt.show()
```
