```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
df = pd.read_csv('./data/suwon_annual_temperature.csv', encoding='cp949', skiprows=6)
```

```python
df
```

```python
df.dropna(axis=0, inplace=True)
```

```python
df
```

```python
df.columns = ['년', '지점', '평균기온', '평균최저기온', '평균최고기온']
```

```python
df
```

```python
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['font.family'] = 'AppleGothic' 
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import numpy as np
```

```python
plt.figure(figsize=(16, 8))
plt.plot(df['년'], df['평균기온'], label='평균기온', color='green', marker='o')
plt.plot(df['년'], df['평균최저기온'], label='평균최저기온', color='blue', marker='o')
plt.plot(df['년'], df['평균최고기온'], label='평균최고기온', color='red', marker='o')
plt.legend(ncol=3)
plt.xticks(np.arange(2000, 2023, 2), rotation=45)
plt.xlabel('년도', loc='right')
plt.ylabel('기온(℃)', loc='top')
plt.title('수원시 연도별 평균기온 (2000~2022년)')
plt.savefig('./result/suwon_annual_temperature.png')
plt.show()
```

```python
df2 = pd.read_csv('./data/suwon_annual_rainfall.csv', encoding='cp949', skiprows=6)
```

```python
df2
```

```python
plt.figure(figsize=(16, 8))
plt.bar(df2['년'], df2['강수량(mm)'], label='강수량', color='skyblue')
plt.xlabel('년도', loc='right')
plt.ylabel('강수량(mm)', loc='top')
plt.yticks(rotation=45)
plt.xticks(np.arange(2000, 2023, 2), rotation=45)
plt.title('수원시 연도별 강수량 (2000~2022년)')
plt.savefig('./result/suwon_annual_rainfall.png')
plt.show()
```
